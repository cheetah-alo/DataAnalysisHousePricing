# Importamos las librerías

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.dummy import DummyClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import *

#from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import models
from tensorflow.keras.constraints import max_norm

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, auc, roc_auc_score
```

<hr>


## <a name="M1">1. Carga de datos</a>

```python
df = pd.read_csv("./Churn_Modelling.csv")
df.sample(5)
```

<hr>


## <a name="M2">2. Preparación de los datos</a>

### <a name="M21">2.1. Análisis del desbalanceo de datos</a>


* En el Dataset hay un 80% de clientes que no abandonan la compañia y un 20% que si.

```python
churn = df['Exited'].value_counts()
print('Numero de Churns:\n{}'.format(churn))
```

### <a name="M22">2.2. Particionado de datos en conjuntos de entrenamiento y test</a>


* Vamos a dividir los datos en conjuntos de entrenamiento y test, seleccionando:
    + 80% de entrenemiento
    + 20% de test
    
    
* Para ello seleccionamos los 8.000 primeros elementos de entrenamiento y los 2.000 último elementos de test.

```python
df_train = df.iloc[:8000,:]
df_test = df.iloc[8000:,:]

print("elementos de entrenamiento {}".format(df_train.shape))
print("elementos de test {}".format(df_test.shape))
```

### <a name="M23">2.3. Transformación de variables categóricas</a>


* Dado que tenemos las variables "Gender" y "Geography" como categóricas y son variables interesantes para el análisis, tenemos que transformarlas a variables numéricas.


* Para ello aplicaremos un:
    + Codificación binaria para Gender
    + One-hot para Geography

```python
# Codificamos la variable Gender
lb = LabelBinarizer()
df_train['Gender'] = lb.fit_transform(df_train['Gender'])
df_test['Gender'] = lb.fit_transform(df_test['Gender'])

# Codificamos la variable geograpy
df_dummy_train = pd.get_dummies(df_train['Geography'])
df_dummy_test = pd.get_dummies(df_test['Geography'])
df_train = pd.concat([df_train, df_dummy_train], axis=1)
df_test = pd.concat([df_test, df_dummy_test], axis=1)
df_train = df_train.drop(['Geography'], 1)
df_test = df_test.drop(['Geography'], 1)

# mostramos un sample de entrenamiento
df_train.sample(5)
```

```python
# mostramos un sample de test
df_test.sample(5)
```

### <a name="M24">2.4. Adición de datos a la clase desbalanceada</a>


* Dado que tenemos un porcentaje de clientes relativamente pequeño que abandonan la compañia, debemos de darles más peso para los algoritmos de aprendizaje. 

```python
NUM_SAMPLES=3000

df_train_class_des = df_train[df_train['Exited']==1].sample(NUM_SAMPLES, replace=True)
df_train_ext = pd.concat([df_train, df_train_class_des])
print("elementos de entrenamiento a añadir {}".format(len(df_train_class_des['Exited']==1)))
print("elementos del dataframe con los datos añadidos {}".format(len(df_train_ext[df_train_ext["Exited"]==0])))
```

### <a name="M25">2.5. Normalización Cambio de estructura de datos a numpy</a>

```python
# Variables a seleccionar
X_features = ['CreditScore', 'Gender', 'Age', 'Tenure', 'Balance', 
               'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
y_feature = ['Exited']

# Dataframes de entrenamiento y test
df_train_ext = df_train_ext[X_features + y_feature]
df_test = df_test[X_features + y_feature]

# Paso de datos a numpy y normalización
MMS = MinMaxScaler()
y_train = df_train_ext['Exited'].values
X_train = df_train_ext.drop(y_feature, 1).values
X_train_norm = MMS.fit_transform(X_train)

y_test = df_test['Exited'].values
X_test = df_test.drop(y_feature, 1).values
X_test_norm = MMS.fit_transform(X_test)
```

<hr>

## <a name="M3">3. Creacción y ajuste de los modelos</a>

* Para la creación y ajuste de los modelos se proponen utilizar los siguientes Algoritmos de Aprendizaje:
    - Clasificador dummy
    - Gaussian Naive Bayes
    - Decision Tree Classifier: Se propone crear un arbol de 5 niveles con criterio Gini
    - Random Forest Classifier: Se propone crear un random forest con 100 estimadores y arboles de decisión de 5 niveles de profundidad
    - Ada Boost: Se propone crear un Ada Boost con 100 Arboles de decisión de profundidad 5
    - Neuronal Network

```python
# Clasificador dummy
dm = DummyClassifier()

# Gaussian Naive Bayes
gnb = GaussianNB()

# Decision Tree con 5 niveles de profundidad
dtc_e = DecisionTreeClassifier(max_depth = 5, criterion = 'entropy', random_state = 1)
dtc_g = DecisionTreeClassifier(max_depth = 5, random_state = 1)

# Random forest con 100 estimadores y arboles de decisión de 5 niveles de profundidad
rfc_e = RandomForestClassifier(random_state=1, n_estimators=100, max_depth = 5, criterion = 'entropy')
rfc_e2 = RandomForestClassifier(random_state=1, n_estimators=300, max_depth = 5, criterion = 'entropy')
rfc_g = RandomForestClassifier(random_state=1, n_estimators=100, max_depth = 5)
rfc_g2 = RandomForestClassifier(random_state=1, n_estimators=300, max_depth = 5)

# Ada Boost con 100 Arboles de decisión de profundidad 5
abc_l1 = AdaBoostClassifier(n_estimators=100, random_state=0)
abc_l08 = AdaBoostClassifier(n_estimators=100, random_state=0, learning_rate=0.8)
abc_l1_est_mod = AdaBoostClassifier(n_estimators=300, random_state=0)
abc_l08_est_mod = AdaBoostClassifier(n_estimators=300, random_state=0, learning_rate=0.8)

# Red neuronal 
model = Sequential()

# input layer
model.add(Dense(120,  activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(80,  activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(40, activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(20, activation='relu'))
model.add(Dropout(0.2))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Red neuronal 2
model2 = Sequential()

# input layer
model2.add(Dense(120,  activation='relu'))
model2.add(Dropout(0.2))

# hidden layer
model2.add(Dense(80,  activation='relu'))
model2.add(Dropout(0.2))

# output layer
model2.add(Dense(units=1,activation='sigmoid'))

# Compile model
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Creamos un diccionario con los clasificadores
clasificadores = {'Dummy': dm,
                  'Gaussian': gnb,
                  'Decision Tree entropy': dtc_e,
                  'Decision Tree gini': dtc_g,
                  'Random Forest entropy': rfc_e,
                  'Random Forest gini': rfc_g,
                  'Random Forest entropy 2': rfc_e2,
                  'Random Forest gini 2': rfc_g2,
                  'Ada Boost 1':abc_l1,
                  'Ada Boost 08':abc_l08,
                  'Ada Boost 1 estimator mod':abc_l1_est_mod,
                  'Ada Boost 08 estimator mod':abc_l08_est_mod,
                  'Red Neuronal': model,
                  'Red Neuronal 2': model2}
```

* Ajustamos los modelos

```python
# Ajustamos los modelos
for k, v in clasificadores.items():
    print ('CREANDO MODELO: {clas}'.format(clas=k))
    if k == "Red Neuronal":
        model.fit(X_train_norm, 
              y_train, 
              validation_split=0.1, 
              epochs=20, 
              batch_size=32, 
              verbose=2)
    elif k == "Red Neuronal 2":
        model2.fit(X_train_norm, 
              y_train, 
              validation_split=0.1, 
              epochs=20, 
              batch_size=32, 
              verbose=2)
    else:
        v.fit(X_train_norm, y_train)
    
```

<hr>

## <a name="M4">4. Evaluación de los modelos</a>

* Se propone evaluar los modelos tanto con el conjunto de datos de entrenamiento como de test con la métricas de:

    + Accuracy
    + Precision
    + Recall
    + F1

```python
evaluacion = list()
for k, v in clasificadores.items():
    print ('EVALUANDO MODELO: {model}'.format(model=k))
    model = {}
    model['name'] = k
    if k == "Red Neuronal" or k =='Red Neuronal 2':
        y_pred_train = v.predict_classes(X_train_norm)
        y_pred_test = v.predict_classes(X_test_norm)
    else:
        y_pred_train = v.predict(X_train_norm)
        y_pred_test = v.predict(X_test_norm)
        
    model['accuracy_train'] = accuracy_score(y_true=y_train, y_pred=y_pred_train)
    model['accuracy_tests'] = accuracy_score(y_true=y_test, y_pred=y_pred_test)
    model['precision_train'] = precision_score(y_true=y_train, y_pred=y_pred_train)
    model['precision_tests'] = precision_score(y_true=y_test, y_pred=y_pred_test)
    model['recall_train'] = recall_score(y_true=y_train, y_pred=y_pred_train)
    model['recall_tests'] = recall_score(y_true=y_test, y_pred=y_pred_test)
    model['f1_train'] = f1_score(y_true=y_train, y_pred=y_pred_train)
    model['f1_tests'] = f1_score(y_true=y_test, y_pred=y_pred_test)
    model['AUC_train'] = roc_auc_score(y_train, y_pred_train)
    model['AUC_test'] = roc_auc_score(y_test, y_pred_test)
    evaluacion.append(model)

# Pasamos los resultados a un DataFrame para visualizarlos mejor
df = pd.DataFrame.from_dict(evaluacion)
df.set_index("name", inplace=True)
df.sort_values('accuracy_tests', ascending=False)
```

### <a name="M41">4.1. Matrices de confusión</a>

* A continuación se pintarán las matrices de confusión para los diferentes modelos generados tanto para los datos de entrenamiento como los datos de test.

```python
def plot_confusion_matrix(cm, classes, title, cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'), horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
    
# Calculamos las matrices de confusión
msc = list()
for k, v in clasificadores.items():
    print ('Obteniendo Matriz de Confusión de: {model}'.format(model=k))
    model = {}
    model['name'] = k
    if k == "Red Neuronal" or k == 'Red Neuronal 2':
        y_pred_train = v.predict_classes(X_train_norm)
        y_pred_test = v.predict_classes(X_test_norm)
    else:
        y_pred_train = v.predict(X_train_norm)
        y_pred_test = v.predict(X_test_norm)
    model['confusion_matrix_train'] = confusion_matrix(y_true=y_train, y_pred=y_pred_train)
    model['confusion_matrix_test'] = confusion_matrix(y_true=y_test, y_pred=y_pred_test)
    msc.append(model)
```

* Pintamos las matrices de confusión con el conjunto de datos de entrenamiento.

```python
fig = plt.figure(figsize=(20, 60))
for i, mc in enumerate(msc):
    ax = plt.subplot(10, 2, i+1)
    sns.set(font_scale=1.5) 
    sns.heatmap(mc['confusion_matrix_train'],annot=True, ax=ax, cmap="Blues", fmt="g")
    #plot_confusion_matrix(mc['confusion_matrix_train'], classes=['0', '1'], 
                          #title='{}\nMatriz de Confusión Datos Entrenamiento'.format(mc['name']))
    label_font = {'size':'13'}
    ax.set_xlabel('Predicted labels', fontdict=label_font);
    ax.set_ylabel('Observed labels', fontdict=label_font);
    title_font = {'size':'10'}
    ax.set_title('{}\nMatriz de Confusión Datos Entrenamiento'.format(mc['name']));
    
fig.tight_layout()
plt.show()
```

* Pintamos las matrices de confusión con el conjunto de datos de test.

```python
fig = plt.figure(figsize=(20, 60))
for i, mc in enumerate(msc):
    ax = plt.subplot(10, 2, i+1)
    sns.set(font_scale=1.5) 
    sns.heatmap(mc['confusion_matrix_test'],annot=True, ax=ax, cmap="Blues", fmt="g")
    #plot_confusion_matrix(mc['confusion_matrix_train'], classes=['0', '1'], 
                          #title='{}\nMatriz de Confusión Datos Entrenamiento'.format(mc['name']))
    label_font = {'size':'13'}
    ax.set_xlabel('Predicted labels', fontdict=label_font);
    ax.set_ylabel('Observed labels', fontdict=label_font);
    title_font = {'size':'10'}
    ax.set_title('{}\nMatriz de Confusión Datos Test'.format(mc['name']));
    
fig.tight_layout()
plt.show()
```

## <a name="M5">5. Conclusiones</a>


Con el falso positivo pensamos que el cliente se va a ir y le ofrecemos alguna promoción, invertimos en ese cliente. En realidad no se iba a ir y estamos invirtiendo cuando no hubiese sido necesario. Sin embargo, una promoción a un cliente ya satisfecho seguramente haga que esté aún más satisfecho y pueda hacer publicidad de nuestro servicio. Por lo que el costo para los falsos positivos no es grave.

Con el falso negativo no estamos detectando que el cliente está descontento, por lo que directamente no le estamos aplicando ninguna oferta o promoción. este coste es mayor porque el cliente se va a ir descontento, además de que podría hacer mala publicidad y dañar la imagen corporativa con lo cuál el coste de este cliente es muy alto, y no vamos a actuar para impedirlo.

Con los verdaderos positivos se podrá detectar aquellos clientes que se quieren ir y se podrá realizar una oferta para retenerlos.

Por último, el verdadero negativo no tiene ningún coste ya que nos informa de los clientes que no tienen intención de marcharse y además han sido detectados correctamente.

Para identificar cuál es el mejor modelo vamos a utilizar la métrica recall que indica el porcentaje de acierto respecto al total de personas que pretender irse, es decir, estamos detectando mejor todos los churn.

Nos hemos basado para elegir esta métrica en intentar identificar el modelo que menos falsos negativos tiene, porque es el cliente que más coste nos puede suponer y esto se consigue detectar con el recall. Cuanto más alto es el recall, más porcentaje de acierto en detectar los churn.

Según el recall, el mejor modelo es el Decision Tree Gini conseguimos identificar casi el 72% de las clientes que quieren irse pero hay alrededor de un 28% de clientes que desean marcharse y no lo estamos detectando
